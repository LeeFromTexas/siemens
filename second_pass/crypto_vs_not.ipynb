```python
import pandas as pd
data = []
data2 = []

with open("crypto_text.txt") as f2:
    data = f2.read().split()
    print(type(data))
    print(data)

with open("astar_text.txt") as f2:
    data2 = f2.read().split()
    print(type(data2))
    print(data2)
```

```python
crypto = pd.DataFrame(data)
astar = pd.DataFrame(data2)
```

```python
crypto['category'] = 'type_a'
astar['category'] = 'type_b'
crypto.head()
```

```python
frames = [crypto,astar]
result = pd.concat(frames)
result.head()
```

```python
result.columns
result.columns = ['text', 'category']
```

```python
import scattertext as st
import spacy
#convention_df = st.SampleCorpora.ConventionData2012.get_data()
corpus = (st.CorpusFromPandas(result,
                           category_col='category',
                           text_col='text',
                        nlp = spacy.load('en_core_web_sm'))
          .build()
          .get_unigram_corpus())
html = st.produce_frequency_explorer(
    corpus,
    category='type_a',
    term_scorer=st.ScaledFScorePresets(one_to_neg_one=True, use_score_difference=True),
    #metadata=convention_df['speaker'],
    grey_threshold=0
)
file_name = 'crypto.html'
open(file_name, 'wb').write(html.encode('utf-8'))
print('Open ./%s in Chrome.' % (file_name))
```
